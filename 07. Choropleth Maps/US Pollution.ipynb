```python
#import the packages
import pandas as pd
import numpy as np 
#%matplotlib inline
import plotly as ps
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot 
init_notebook_mode(connected=True) 
```

```python
exportData = pd.read_csv('C:/Users/rarora17/Desktop/Hackathon/datacsv_V1_Pollution.csv')
exportData.head()
```

```python
Year2011 = exportData.loc[exportData['Year'] == 2011]

plot2011_Pollution = dict(type ='choropleth',
           locations = exportData['State Code'],
           locationmode = 'USA-states',
           colorscale = 'YlOrRd_r',
           text=Year2011['NO2'],
           z = Year2011['NO2'],
            marker = dict(line=dict(color='rgb(255,255,255)',width=2)),
           colorbar = {'title':'Millions USD'})

myLayout_2011 = dict( geo=dict(scope='usa',showlakes=True,lakecolor='rgb(85,173,240)'),
                title='Exportd Per Us State')
myChoroMap_2011 = go.Figure(data=[plot2011_Pollution],layout=myLayout_2011)
iplot(myChoroMap_2011)
```
