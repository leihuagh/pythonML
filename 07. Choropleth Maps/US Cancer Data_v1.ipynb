```python
#import the packages
import pandas as pd
import numpy as np 
#%matplotlib inline
import plotly as ps
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot 
init_notebook_mode(connected=True) 
```

```python
exportData = pd.read_csv('C:/Users/rarora17/Desktop/Hackathon/CancerRaw_Data.csv')
exportData.head()
```

```python
Year2011 = exportData.loc[exportData['YEAR'] == 2011]
Year2011.tail()
```

```python
Year2011 = exportData.loc[exportData['YEAR'] == 2011]

plot2011_Pollution = dict(type ='choropleth',
           locations = exportData['StateCode'],
           locationmode = 'USA-states',
           colorscale = 'YlOrRd_r',
           text=Year2011['Data_Value'],
           z = Year2011['Lung Cancer'],
           marker = dict(line=dict(color='rgb(255,255,255)',width=1)),
           colorbar = {'title':'Cancer Rates'})
myLayout_2011 = dict( geo=dict(scope='usa',showlakes=True,lakecolor='rgb(85,173,240)'),
                title='Cancer Rates in states in Year 2011')
myChoroMap_2011 = go.Figure(data=[plot2011_Pollution],layout=myLayout_2011)
iplot(myChoroMap_2011)
```
