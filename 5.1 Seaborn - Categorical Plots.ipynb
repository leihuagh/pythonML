```python
# import the 'seaborn' package
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Draw the distribution plot
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

 # Pair Plots
     
'pairplot' are used to draw a plt of all the pairs from the selected data. The Diagonal shows the distribution of the data for the column/row it represents. 

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## RugPlot

```python
sns.rugplot(tips['total_bill'])
```

## Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Count Plot

```python
sns.countplot(x='sex',data=tips)
```

### Box Plots 

Whisker plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
myplot=sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plots

```python
myplot=sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
myplot=sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python

myplot=sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)

```

### Strip Plot

```python
myplot=sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
myplot=sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
myplot=sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='smoker',dodge=True)
```

### Swarm Plot

```python
myplot=sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker')
#                     ,split=True)
```

## Two Plots in one 

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

### Factor Plot

```python
sns.factorplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```
